### Importing the 4 datasets (job_postings, job_skills,company_industries,company_name)

```python
job_postings = pd.read_csv(r'/Users/pierremora/Desktop/LinkedIn Job Postings 2023/DATA/CLEAN DATA/job_postings_clean.csv', index_col = False)
```

```python
job_skills = pd.read_csv(r'/Users/pierremora/Desktop/LinkedIn Job Postings 2023/DATA/CLEAN DATA/job_skills_clean.csv', index_col = False)
```

```python
company_industries = pd.read_csv(r'/Users/pierremora/Desktop/LinkedIn Job Postings 2023/DATA/CLEAN DATA/company_industries_clean.csv', index_col = False)
```

```python
company_name = pd.read_csv(r'/Users/pierremora/Desktop/LinkedIn Job Postings 2023/DATA/CLEAN DATA/company_name_clean.csv', index_col = False)
```

### Merging job_postings dataframe into company_industries dataframe

```python
# Merging the datasets on the 'company_id' column
merged_data1 = pd.merge(job_postings, company_industries, on="company_id", how="left")
```

```python
#checking results
merged_data1.dtypes
```

```python
#checking dataframe head
merged_data1.head()
```

### dropping unnecesary columns 

```python
# Dropping the 'Unnamed: 0_x' and 'Unnamed: 0_y' columns
merged_data1= merged_data1.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'])
```

```python
merged_data1.head(1)
```

```python
# Dropping rows where the 'company_id' value is missing (NaN)
merged_data1_clean = merged_data1.dropna(subset=['company_id'])
```

### Merging the actual main dataset with company_name dataframe

```python
# Merging the datasets on the 'company_id' column
merged_data2 = pd.merge(merged_data1_clean, company_name, on="company_id", how="left")
```
